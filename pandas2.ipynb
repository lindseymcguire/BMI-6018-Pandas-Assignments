```python
# Question 3 (15 Points)
# Change the order of columns of a dataframe. Create a generic function to interchange two columns, without
# hardcoding column names.
# Input
df = pd.DataFrame(np.arange(20).reshape(-1, 5), columns=list('abcde'))
# making a function where the paramaters are the dataframe, and the two cols you want to interchange
def change_columns(dataframe, col1, col2):
    # getting a list of the columns in a dataframe
    cols = list(dataframe.columns)
    # getting the index of each column
    index_col1 = cols.index(col1)
    index_col2 = cols.index(col2)
    # assigning the index of one column to the other
    cols[index_col1], cols[index_col2] = cols[index_col2], cols[index_col1]
    return dataframe[cols]

changed_columns_df = change_columns(df, 'a', 'e')
changed_columns_df
```
